```python
import pandas as pd
import numpy as np
from scipy import stats
file_to_load = "belize_airbnb_listings.csv"
df = pd.read_csv(file_to_load)
df1 = pd.DataFrame(df)
df1
```

# Arithmetic mean


Arithmetic mean is sum of the values divided by the total number of values.

```python
df1[["price","reviews_per_month","availability_365"]].mean()
```

# Geomertic mean

```python
geometric mean is the N-th root of the product of all values, where N is the number of values.
```

```python
print(stats.gmean(df1["minimum_nights"]))
```

# Harmonic mean

```python
Harmonic mean is number of values divides by the sum of the reciprocal of the values.
```

```python
print(stats.hmean(df1["price"]))
```
